```python
plt.figure(figsize=(10, 6))
plt.hist(df['co2'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of CO₂ Emissions')
plt.xlabel('CO₂ Emissions (g/mi)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['displ'], bins=15, color='plum', edgecolor='black')
plt.title('Engine Displacement Distribution')
plt.xlabel('Engine Displacement (L)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['pv2'].dropna(), bins=20, color='orange', edgecolor='black')
plt.title('Histogram of pv2')
plt.xlabel('pv2')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Compute correlation matrix
corr = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr, cmap='RdBu_r', annot=False, center=0)
plt.title('Correlation Heatmap of Fuel Economy Dataset')
plt.show()
```

```python
## Interpret the data - What story does the visualization tell?
#
# The heatmap illustrates how numerical features in the fuel economy dataset are interrelated.
# `ghgScore` has a strong negative correlation with `co2`, indicating that as CO2 emissions increase, the greenhouse gas score (efficiency) decreases.
# `ghgScore` is strongly positively correlated with `feScore`, `comb`, and other MPG metrics such as `city`, `highway`, `UCity`, and `UHighway`, suggesting that better fuel economy is associated with a higher environmental score.
# Engine-related variables like `displ` (engine displacement), `cylinders`, and `pv2` show moderate to strong negative correlations with `ghgScore`, implying that vehicles with larger engines and more power tend to have lower efficiency.
# `id` and `year` have minimal or no correlation with `ghgScore`, suggesting they do not contribute meaningfully to predicting the target.
```
